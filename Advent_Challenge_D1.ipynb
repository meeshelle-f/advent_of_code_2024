```python
import numpy as np
import pandas as pd


def replace_non_integers_with_zero(df):
  """Replaces non-integer values in a DataFrame with 0.

  Args:
    df: The input DataFrame.

  Returns:
    The DataFrame with non-integer values replaced by 0.
  """

  for col in df.columns:
    df[col] = df[col].apply(lambda x: 0 if not isinstance(x, (int, np.integer)) else x)

  return df

df = replace_non_integers_with_zero(df)
df
```
