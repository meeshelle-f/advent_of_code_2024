**create example function**

identify all the ways XMAS can be spelled out (vertical, horizontal, direction of letters does not matter!)

In the example XMAS can be spelled out 18 times

```python
import re

# CASE 1: left-to-right horizontal
# CASE 2: right-to-left horizontal
# CASE 3: Check up-down vertical
# CASE 4: Check down-up vertical
# CASE 5: Diagonol upleft-to-downright
# CASE 6: Diagonol downright-to-upleft
# CASE 7: Diagonol upright-to-downleft
# CASE 8: Diagonol downleft-to-upright #1,9

def save_xmas(df):
    """Finds all occurrences of "XMAS" in a given DataFrame.

    Args:
        df: The input DataFrame.

    Returns:
        A list of tuples, each containing the starting row, starting column, and direction of the "XMAS" pattern.
    """
    patterns =  []
    rows, cols = df.shape # 10 10

    #iterate through all rows and columns within the crossword and save xmas!!!
    for row in range(rows):
      # print(row) #0- 9
      for col in range(cols):
        #is iterating through the first row, column-by-column
        # print(row, col, df.iloc[row,col] ) #0 0 M, 0 1 M, 0 2 M, 0 3 S
        # print(row, col, df.iloc[row,col])
        if df.iloc[row,col] == 'X':  #0 5 X
          # print(col + 3) #8
          # print(cols) #10

          # CASE 1: left-to-right horizontal
          #if X + 3 moving to the right fits within the df, and is followed by MAS...
          if col + 3 < cols and df.iloc[row, col+1] == 'M' and df.iloc[row, col+2] == 'A' and df.iloc[row, col+3] == 'S':
            patterns.append((row, col, 'left-to-right horizontal')) # captures the starting point of crossword 'X' of 0 5 X
          # CASE 2: right-to-left horizontal
          if col - 3 >= 0 and df.iloc[row, col-1] == 'M' and df.iloc[row, col-2] == 'A' and df.iloc[row, col-3] == 'S':
            patterns.append((row, col, 'right-to-left horizontal')) # captures the starting point of crossword 'X' of 0 5 X

          # CASE 3: Check up-down vertical
          if row + 3 < rows and df.iloc[row+1, col] == 'M' and df.iloc[row+2, col] == 'A' and df.iloc[row+3, col] == 'S':
            patterns.append((row, col, 'up-down vertical')) # captures the starting point of crossword 'X' of 0 5 X

          # CASE 4: Check down-up vertical
          if row - 3 >= 0 and df.iloc[row-1, col] == 'M' and df.iloc[row-2, col] == 'A' and df.iloc[row-3, col] == 'S':
            patterns.append((row, col, 'down-up vertical')) # captures the starting point of crossword 'X' of 0 5 X

          # CASE 5: Diagonol upleft-to-downright
          if row + 3 < rows and col + 3 < cols and df.iloc[row+1, col+1] == 'M' and df.iloc[row+2, col+2] == 'A' and df.iloc[row+3, col+3] == 'S':
            patterns.append((row, col, 'upleft-to-downright diagonal')) # captures the starting point of crossword 'X' of 0 5 X

          # CASE 6: Diagonol downright-to-upleft
          if row - 3 >= 0 and col - 3 >= 0 and df.iloc[row-1, col-1] == 'M' and df.iloc[row-2, col-2] == 'A' and df.iloc[row-3, col-3] == 'S':
            patterns.append((row, col, 'downright-to-upleft diagonal')) # captures the starting point of crossword 'X' of 0 5 X

          # CASE 7: Diagonol upright-to-downleft where rows are increasing but cols decrease
          if row + 3 < rows and col - 3 >= 0 and df.iloc[row+1, col-1] == 'M' and df.iloc[row+2, col-2] == 'A' and df.iloc[row+3, col-3] == 'S':
            patterns.append((row, col, 'upright-to-downleft diagonal')) # captures the starting point of crossword 'X' of 0 5 X

          # CASE 8: Diagonol downleft-to-upright where rows are decreasing but cols increase
          if row - 3 >= 0  and col + 3 < cols and df.iloc[row-1, col+1] == 'M' and df.iloc[row-2, col+2] == 'A' and df.iloc[row-3, col+3] == 'S':
            patterns.append((row, col, 'downleft-to-upright diagonal')) # captures the starting point of crossword 'X' of 0 5 X


    # print(rows,cols)
    return patterns

patterns = save_xmas(df)
i=0
for pattern in patterns:
    i+=1
    row, col, direction = pattern
    print(f"{i} Found 'XMAS' at ({row}, {col}), direction: {direction}")
```

Looking for the instructions, you flip over the word search to find that this isn't actually an XMAS puzzle; it's an X-MAS puzzle in which you're supposed to find two MAS in the shape of an X. One way to achieve that is like this:

```python
##example data
import pandas as pd

data = """
.M.S......
..A..MSMS.
.M.S.MAA..
..A.ASMSM.
.M.S.M....
..........
S.S.S.S.S.
.A.A.A.A..
M.M.M.M.M.
..........
"""

# Create a list of lists, where each inner list represents a row of the data
data_list = [list(line) for line in data.strip().split('\n')]

# Create a DataFrame from the list of lists
df = pd.DataFrame(data_list)

print(df) #2MAS in X shape appear 9 times...
```

```python
import re


def save_xmas(df):
    """Finds all occurrences of "XMAS" in a given DataFrame.

    Args:
        df: The input DataFrame.

    Returns:
        A list of tuples, each containing the starting row, starting column, and direction of the "XMAS" pattern.
    """
    patterns =  []
    rows, cols = df.shape # 10 10

    #iterate through all rows and columns within the crossword and save xmas!!!
    for row in range(rows):
      for col in range(cols):

        if df.iloc[row,col] == 'A':  #0 5 X

          if row - 1 >= 0  and col - 1 >= 0 and row + 1 < rows and col + 1 < cols:

            # CASE 1:
            #  0 1 2
            #0 M   M
            #1   A
            #2 S   S

            if df.iloc[row-1, col-1] == 'M'  and df.iloc[row-1, col+1] == 'M' and df.iloc[row+1, col+1] == 'S' and df.iloc[row+1, col-1] == 'S':
              patterns.append((row, col)) # captures the starting point of crossword 'X' of 0 5 X

            # CASE 2:
            #  0 1 2
            #0 S   S
            #1   A
            #2 M   M
            if df.iloc[row-1, col-1] == 'S'  and df.iloc[row-1, col+1] == 'S' and df.iloc[row+1, col+1] == 'M' and df.iloc[row+1, col-1] == 'M':
              patterns.append((row, col)) # captures the starting point of crossword 'X' of 0 5 X


            # CASE 3: SS in same col 1 switch
            #  0 1 2
            #0 S   M
            #1   A
            #2 S   M

            if df.iloc[row-1, col-1] == 'S'  and df.iloc[row-1, col+1] == 'M' and df.iloc[row+1, col+1] == 'M' and df.iloc[row+1, col-1] == 'S':
              patterns.append((row, col)) # captures the starting point of crossword 'X' of 0 5 X

            # CASE 4: SS in same col 1 switch
            #  0 1 2
            #0 M   S
            #1   A
            #2 M   S
            if df.iloc[row-1, col-1] == 'M'  and df.iloc[row-1, col+1] == 'S' and df.iloc[row+1, col+1] == 'S' and df.iloc[row+1, col-1] == 'M':
              patterns.append((row, col)) # captures the starting point of crossword 'X' of 0 5 X

    return patterns

patterns = save_xmas(df)
i=0
for pattern in patterns:
    i+=1
    row, col = pattern
    print(f"{i} Found 'XMAS' at ({row}, {col})")
```
